```python
import pandas as pd
from scipy.stats import pearsonr

# Load your data
df = pd.read_excel("Data.xlsx")

# Clean and prepare
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True, errors='coerce')
df = df.dropna(subset=["Date"])
df["Story_Posted_Binary"] = (df["Story_Posted"] > 0).astype(int)

# Point-Biserial Correlation = Pearson between numeric and binary variable
r_value, p_value = pearsonr(df["Calorie_Output"], df["Story_Posted_Binary"])

# Display results
print("ğŸ“ Hypothesis Testing Result:")
print(f"Correlation Coefficient (r): {r_value:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("âœ… Statistically significant â€” reject the null hypothesis.")
else:
    print("âŒ Not statistically significant â€” fail to reject the null hypothesis.")
```
